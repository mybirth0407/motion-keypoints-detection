```python
import numpy as np
import pandas as pd
import os

import cv2
import matplotlib.pyplot as plt
```

```python
def vis_keypoints(data_dir, data, n=3, color=(255, 0, 0), diameter=5):
    df = pd.read_csv(data)
    for i in range(n):
        rand = np.random.randint(len(df))

        filename = df.iloc[rand, 0]
        preds = df.iloc[rand, 1:].to_numpy()

        img = cv2.imread(os.path.join(data_dir, filename))
        a_keypoints = []
        for i in range(0, preds.shape[0], 2):
            a_keypoints.append((float(preds[i]), float(preds[i+1])))

        for (x, y) in a_keypoints:
            cv2.circle(img, (int(x), int(y)), diameter, color, -1)

        plt.figure(figsize=(10, 10))
        plt.axis('off')
        plt.imshow(img)
```

```python
vis_keypoints('data/test_imgs', 'submissions/submissions.csv')
```

```python
vis_keypoints('data/train_imgs', 'data/train_df_modified.csv')
```

```python
vis_keypoints('data/cropped_train_imgs', 'data/cropped_train_df.csv')
```
